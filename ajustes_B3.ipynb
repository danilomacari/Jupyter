```python
import csv
import requests
from bs4 import BeautifulSoup
ender = 'http://www2.bmf.com.br/pages/portal/bmfbovespa/lumis/lum-ajustes-do-pregao-ptBR.asp'
download = requests.get(ender)
soup = BeautifulSoup(download.content, 'html.parser')
tabela=soup.find_all('td')
S=[item.text.split('</td>,')[0] for item in tabela]
lista=[]
for item in S:
    if len(item)>50:
        ativo=item[:3]
        listaAux=[ativo]
    elif len(item)==0:
        lista.append(listaAux)
        listaAux=[ativo]
    else:
        correctValue=item.replace('.','')
        correctValue=correctValue.replace(',','.')
        listaAux.append(correctValue.strip())
lista.append(listaAux)
filename='ajustes.csv'
with open(filename, 'w') as myfile:
    wr = csv.writer(myfile,quoting=csv.QUOTE_NONE)
    for item in lista:
        wr.writerow(item)
    
lista
```
